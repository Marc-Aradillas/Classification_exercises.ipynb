We can read the results of a SQL query into a dataframe using pd.read_sql.

```python
db_url = get_connection('chipotle')

query = '''
        SELECT *
        FROM orders
        '''
```

```python
chip = pd.read_sql(query, db_url)
chip.head()
```

Let's store the results of this SQL query in a csv file!

```python
chip.to_csv('chip.csv', index = 0)
```

We can read back in our data from the csv file instead of the SQL query.

```python
chip_local = pd.read_csv('chip.csv')
chip_local.head()
```

```python
# Develop a function for saving in your machine

def get_chip():

    filename = 'chip.csv'

    if os.path.isfile(filename):
    
        print('Found your sauce dawg!')
        
        return pd.read_csv(filename)
        
    else:
        
        print('Retieving da sauce!!!')

        url = get_connection('chipotle')
    
        query = '''
                SELECT *
                FROM orders
                '''
        
        chip = pd.read_sql(query, url)
        chip.to_csv(filename, index = 0)
    
        return chip
```

```python
first_chip = get_chip()
first_chip.head()
```

```python
second_chip = get_chip()
second_chip.head()
```
